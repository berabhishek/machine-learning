So the goal of this notebook is to understand more about *Dimensionality Reduction* using PCA.  
The dataset being used is the *iris dataset* from the sklearn package and the only library function used is to calculate the eigenvectors and the eigenvalues.  

```python
from sklearn.datasets import load_iris
import numpy as np
from matplotlib import pyplot as plt
```

```python
iris_data = load_iris()
```

## About the iris dataset
The [iris dataset](https://en.wikipedia.org/wiki/Iris_flower_data_set) contains information on the sepal and petal lengths and widths of 3 types of flowers. *(Setosa, Versicolor and Virginica)*.  
There are total **150 samples** and **4 dimensions** to the dataset. Our aim is to reduce this dataset to the minimum of *2-D* and visualize the dataset. 

But first let's try to visualize the data in all 4 dimensions. 

```python
fig = plt.figure()
fig.set_figheight(20)
fig.set_figwidth(20)
for i in range(4):
    ax = fig.add_subplot("22"+str(i+1))
    for j,color in zip(range(0, 150, 50), ["r","g","b"]):
        ax.hist(iris_data["data"][:,i][j:j+50], alpha = 0.5, color = color, label = iris_data["target_names"][int(j/50)])
        ax.legend()
    plt.title(iris_data["feature_names"][i])
    axes = plt.gca()
    axes.set_xlim([0,10])
    axes.set_ylim([0,10])


plt.show()
```

## Getting the Covariance matrix
The second important step is to get the covariance matrix. Due to the nature of the problem this matrix is symmetric across its primary diagonal. To generate the covariance matrix, first of all, the data (every column) is centered around the mean. Then all the dimensions are multiplied with each other in order to generate the covariance matrix.

The function *get_cov_matrix* inputs the data and outputs the final matrix which is (dim X dim) in dimensions, where dim is the number of dimensions. In this case it is 4.

Logically, a positive correlation means increase in the first one also shows increase in the second one, whereas a negative correlation shows the opposite. Following this definition and the table given above, one simple observation that could be made is increase in *sepal width* shows decrease in the remaining 3 parameters. Other parameters show -more or less- a positive relation with other parameters. However, in later sections we'll try to dig into some more hidden patterns.

```python
def get_cov_matrix(iris_data):
    iris_shape = iris_data["data"].shape
    cov_matrix = np.zeros((iris_shape[1], iris_shape[1]))
    iris_data_centered = np.apply_along_axis(lambda x: x - np.mean(x), 0, iris_data["data"])
    cov_matrix = np.dot(iris_data_centered.T, iris_data_centered) / iris_shape[0]
    return cov_matrix, iris_data_centered
            
            
cov_matrix, iris_data_centered = get_cov_matrix(iris_data)
print(cov_matrix)
```

## Getting the EigenValues and EigenVectors
In order to determine the places where the data varies the most we need to find out the eigenvectors and eigenvalues of the matrix. Since, 4X4 matrix is a small matrix, however it is big enough to go through some iterations in order to get unit distance eigenvectors. I am using a library function for this purpose.

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print(eigenvalues, eigenvectors)
```

## Explained Variance
The next important question is how many eigenvectors and eigenvalues to choose to explain the most of the data.  
The Solution is to choose the vectors that captures the maximum proportion variance of the data. This technique is called *"Explained Variance"*. 

```python
variance_list = [(i/sum(eigenvalues))*100 for i in list(sorted(eigenvalues, reverse = True))]
plt.bar(range(1,len(variance_list)+1), variance_list)
plt.xlabel("Principal Components")
plt.ylabel("% Variance Explained")
plt.title("Variance Explained by Principal Components")
plt.xticks([1,2,3,4], ["PC1","PC2","PC3","PC4"])
plt.show()
```

So, it can be seen from the diagram, one eigenvector expresses much of the variance in the data. Together with the first and the second eigenvalues we can precisely capture *97.6%* variance in the data. Hence, we will discard the third and the fourth data. 

```python
#Creating a tuple of eigenvalues and eigenvectors, sorting and picking top 2
eigen_pairs = [(eigenvalues[i], eigenvectors[i])for i in range(len(eigenvalues))]
eigen_pairs = sorted(eigen_pairs, key = lambda x: x[0], reverse = True)
eigen_pairs = eigen_pairs[:2]
eigenvectors_selected = np.vstack([eigen_pairs[0][1], eigen_pairs[1][1]]).T
print(eigenvectors_selected)
```

## Projecting the 4-D data into 2-D space
The first step is to convert the data from a 150 X 4 matrix into our 150 X 2 dataset.  
final_projection(150, 2) = iris_data_centered(150, 4) X eigenvectors_selected(4,2)  
Finally, the data is plotted using a scatterplot

```python
final_projection = np.dot(iris_data_centered, eigenvectors_selected)
print(final_projection.shape)
```

```python
setosa_scatter = plt.scatter(final_projection[:,0][:50], final_projection[:,1][:50], color = "r", alpha = 0.7)
versicolor_scatter = plt.scatter(final_projection[:,0][50:100], final_projection[:,1][50:100], color = "b", alpha = 0.7)
virginica_scatter = plt.scatter(final_projection[:,0][100:], final_projection[:,1][100:], color = "g", alpha = 0.7)
plt.legend((setosa_scatter, versicolor_scatter, virginica_scatter),("Setosa","Versicolor","Virginica"))
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("2-D Projection")
plt.show()
```

# Conclusion
A clear separation could be seen between *Setosa* and *Versicolor and Virginica*. So the technique really captured the variance in 2 dimensions. Hence this technique is a **Dimensionality Reduction** technique. However, the Versicolor and Virginica are still mixed up. Some more detailed processing needs to be done to bring a little separation. But really moving them apart is a very difficult task.
